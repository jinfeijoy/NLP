```python
import pandas as pd
url = 'https://raw.githubusercontent.com/Vijayvj1/twitter-sentiment-analysis-1/master/train.csv'
df = pd.read_csv(url, encoding='latin-1')
df.head(5)
```

```python
df = df.head(10000)
```

Reference: https://medium.com/analytics-vidhya/sentiment-analysis-in-five-steps-using-automl-d16feeab2a36#94dc

```python
from sklearn.model_selection import train_test_split
from autoviml.Auto_NLP import Auto_NLP
train, test = train_test_split(df, test_size=0.2)
```

```python
input_feature, target = "SentimentText", "Sentiment"
```

```python
train_x, test_x, final, predicted= Auto_NLP(input_feature,
                                            train, 
                                            test,
                                            target,
                                            score_type="balanced_accuracy",
                                            top_num_features=100,
                                            modeltype="Classification", 
                                            verbose=2, 
                                            build_model=True)
```

```python
final.predict(test_x[input_feature])
```
