```python
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36'
}

url = 'https://inshorts.com/en/ajax/more_news'
news_offset = 'rlp9r65s-1'
n_loop = 6
```

```python
def print_headlines(response_text):
    title_list = []
    date_list = []
    content_list = []

    soup = BeautifulSoup(response_text, 'lxml')
    headlines = soup.find_all(attrs={"itemprop": "headline"})
    dates = soup.find_all(attrs = {"clas":"date"})
    articles = soup.find_all(attrs = {"itemprop": "articleBody"})   
    for headline in headlines:
        title_list.append(headline.text)
    for date in dates:
        date_list.append(date.text)
    for article in articles:
        content_list.append(article.text)

    news_list = pd.DataFrame(
    {'title': title_list,
     'date': date_list,
     'content': content_list
    })
    return news_list

## Generate Samples:
# url = 'https://inshorts.com/en/read/world'
# response = requests.get(url, headers=headers)
# test = print_headlines(response.text)

# url = 'https://inshorts.com/en/ajax/more_news'
# news_offset = 'rlp9r65s-1'
# n_loop = 6

## Get today's news 
x = 0
while True:
    response = requests.post(url, data={"category": "", "news_offset": news_offset})
    if x > n_loop:
        print('Break in loop: ', x)
        break

    response_json = json.loads(response.text)
       
    subdataset = print_headlines(response_json["html"])
    if x == 0:
        final_data = subdataset.copy()
    else:
        final_data = pd.concat([final_data, subdataset])
    
    news_offset = response_json["min_news_id"]
    x += 1
    
## Check how many news obtained today
today = date.today()
td = today.strftime("%d %b %Y")
weekday = calendar.day_name[today.weekday()]
date_string = td + ',' + weekday
print('--------------Today Date---------------------')
print(date_string)
print('--------------Final Data---------------------')
print(final_data.date.value_counts())

for check in range(1):
    if len(final_data.date.value_counts()) < 2:
        print('Error Message: Please increase loop to include more news for today')

print('--------------Today News---------------------')
final_data = final_data.reset_index()
today_news = final_data[final_data.date == date_string]
print(today_news.date.value_counts())
print('--------------News Sample---------------------')
display(today_news.sample(3))
```

```python
today_news['index'] = today_news.index
today_news['title_tokens'] = textClean.pipeline(today_news['title'].to_list(), multi_gram = [1], lower_case=True, 
                                              deacc=False, encoding='utf8', errors='strict', stem_lemma = 'lemma', 
                                              tag_drop = [], 
                                              nltk_stop=True, 
                                              stop_word_list=[], 
                                              check_numbers=False, word_length=2, remove_consecutives=True)
# today_news.sample(3)
newstokens = today_news.title_tokens.to_list()
top_n_keywords = tfidf.get_top_n_tfidf_bow(today_news.title_tokens.to_list(), no_below = 2,top_n_tokens = 50)
print(top_n_keywords)
DataExploration.generate_word_cloud(newstokens)
lda_allbow, bow_corpus, dictionary = lda.fit_lda(newstokens,no_below = 1, num_topics = 10)
lda.lda_topics(lda_allbow)
```
